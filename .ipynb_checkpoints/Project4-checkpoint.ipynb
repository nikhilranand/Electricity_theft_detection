```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Create Dataframe and Read the dataset using Pandas
dataset = pd.read_csv('meter1a.csv')
dataset.head()
```

```python
#Convert list into Dataframe
df = pd.DataFrame(dataset)
df
```

```python
#assign Normal instance as 0 and theft as 1
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['theft']=encoder.fit_transform(df['theft'])
```

```python
df.theft.unique()
```

```python
df
```

```python

df1=df.drop(['0','Class'],axis=1)
df1
```

```python
## independent and dependent feature
X=df1.drop(labels=['theft'],axis=1)
y=df1.theft
```

```python
X.head()
```

```python
y.head()
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

from sklearn.metrics import accuracy_score,classification_report
```

```python
## Hypeparameter Tuning
params={'max_depth':[3,5,10,None],
              'n_estimators':[100,200,300],
               'criterion':['gini','entropy']
              }
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classfier=RandomForestClassifier()
```

```python
cv=RandomizedSearchCV(classfier,param_distributions=params,scoring='accuracy',cv=5,verbose=3)
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
## prediction
y_pred=cv.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
