```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Create Dataframe and Read the dataset using Pandas
dataset = pd.read_csv('meter1a.csv')
dataset.head()
```

```python
#Convert list into Dataframe
df = pd.DataFrame(dataset)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
# Check Unique values of target variable
df.theft.value_counts()
```

```python
#Droping class feature 
df1 = df.drop(['0','Class'], axis=1)
df1
```

```python
df1
```

```python
# Encoding Normal as 0 and theft1 as 1
df1['theft']= np.where(df1['theft']== 'Normal',0,1)
```

```python
# Check counts
df1.theft.value_counts()
```

## Visualization of Distribution

```python
# PLot density plot for all features
plt.style.use('seaborn')
df1.hist(bins=50, figsize=(20,15), ec = 'b')
plt.show()
```

```python
#Droping theft user
normaluser_df = df1[df1['theft'] == 0]
normaluser_df
```

```python
#Droping normal user
theftuser_df = df1[df1['theft'] == 1]
theftuser_df
```

```python
# Set the style for Seaborn plots
sns.set(style="whitegrid")

# Create subplots for visualizations
fig, axes = plt.subplots(2, 1, figsize=(8, 8))
fig.suptitle("Exploratory Data Visualizations for Normal Users", fontsize=16)

# Plot 1
sns.histplot(data=normaluser_df, x='Electricity:Facility [kW](Hourly)', kde=True, ax=axes[0])
axes[0].set_title("Electricity:Facility [kW](Hourly)")

# Plot 2
sns.boxplot(data= normaluser_df, x='Electricity:Facility [kW](Hourly)', ax=axes[1])
axes[1].set_title("Electricity:Facility [kW](Hourly)")

# Plot 3
#sns.countplot(data=normaluser_df, x='Electricity:Facility [kW](Hourly)', ax=axes[2])
#axes[2].set_title("Electricity:Facility [kW](Hourly)")


# Adjust subplot layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plots
plt.show()
```

```python
# Percentage for PieChart
percentage = df.theft.value_counts(normalize=True)*100
percentage
```

```python
#plotting PieChart
Theftlabels = ["NORMAL", "THEFT"]
plt.figure(figsize =(12, 7))
plt.pie(percentage,labels = Theftlabels,autopct='%1.1f%%')  
plt.title ("Pie Chart of theft", fontsize = 15)
plt.show() 
```
